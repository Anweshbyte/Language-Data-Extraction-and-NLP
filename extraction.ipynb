```python
import os
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Create a directory to store the text files
directory = 'article_text_files'
os.makedirs(directory, exist_ok=True)

# Load the XLSX file
xlsx_file = 'input.xlsx'
df = pd.read_excel(xlsx_file)

# Iterate over each row in the dataframe
for index, row in df.iterrows():
    url_id = str(row['URL_ID'])
    url = str(row['URL'])

    # Send a GET request to the URL
    response = requests.get(url)
    if response.status_code == 200:
        # Parse the HTML content
        soup = BeautifulSoup(response.content, 'html.parser')

        article_title = soup.find('h1').get_text()

        # Find the article text within <p> tags
        paragraphs = soup.find_all('p')

        # Extract the article text
        article_text = '\n'.join([p.get_text() for p in paragraphs])
        article_text = article_text[853:-396]
        # article_text = article_text[:]

        # Save the extracted article in a text file
        file_name = f'{url_id}.txt'
        file_path = os.path.join(directory, file_name)
        with open(file_path, 'w', encoding='utf-8') as file:
            file.write(article_title + '\n')
            file.write(article_text)
            print(f'Article extracted from URL_ID {url_id} and saved in {file_path}')
    else:
        print(f'Failed to fetch URL_ID {url_id} from {url}')
```

```python
import os

directory = "StopWords"
word_list = []

# Iterate over each file in the directory
for filename in os.listdir(directory):
    if filename.endswith(".txt"):
        filepath = os.path.join(directory, filename)
        
        # Read the contents of the file
        with open(filepath, "r", encoding='latin1') as file:
            words = file.read().split("\n")
            
            # Extend the word_list with the words from the file
            word_list.extend(words)

# Remove any empty strings from the word_list
StopWords = list(filter(None, word_list))

# Print the resulting word_list
print(StopWords)
```

```python
StopWords = [word.lower() for word in StopWords]
StopWords
```

```python
import os

# Directory paths
input_directory = 'article_text_files'
output_directory = 'article_without_stopwords'

# Create the output directory if it doesn't exist
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Iterate through each file in the input directory
for filename in os.listdir(input_directory):
    if filename.endswith(".txt"):
        input_filepath = os.path.join(input_directory, filename)
        output_filepath = os.path.join(output_directory, filename)
        with open(input_filepath, 'r') as input_file:
            text = input_file.read()
        
        # Remove stop words from the text
        words = text.split()
        filtered_words = [word for word in words if word.lower() not in StopWords]
        filtered_text = ' '.join(filtered_words)
        
        # Save the filtered text to the output directory
        with open(output_filepath, 'w') as output_file:
            output_file.write(filtered_text)
```

```python
dictionary_directory = 'MasterDictionary'

# Initialize positive and negative word dictionaries
positive_words = {}
negative_words = {}

# Process positive words file
positive_file = os.path.join(dictionary_directory, 'positive-words.txt')
with open(positive_file, 'r' , encoding='latin1' ) as file:
    for line in file:
        word = line.strip().lower()
        if word not in StopWords:
            positive_words[word] = True

# Process negative words file
negative_file = os.path.join(dictionary_directory, 'negative-words.txt')
with open(negative_file, 'r', encoding='latin1') as file:
    for line in file:
        word = line.strip().lower()
        if word not in StopWords:
            negative_words[word] = True

# Print the positive and negative word dictionaries
print("Positive Words:")
for word in positive_words:
    print(word)

print("\nNegative Words:")
for word in negative_words:
    print(word)
```
